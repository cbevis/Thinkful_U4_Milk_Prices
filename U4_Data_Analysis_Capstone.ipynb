```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime
% matplotlib inline
```

```python
#monthly milk production in millions of pounds
header = ['Year', 'Month', 'Sample_Prod', 'Num_States', 'Est_USA_Prod']
production = pd.read_csv('Milk_Prod/Milk_Production_Estimates-Table 1.csv')
production.columns = header
production.infer_objects()
print(production.head())
print(production.dtypes)
```

```python
us_pop = pd.read_csv('US_Pop-Table 1.csv')
us_pop.drop(columns=['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], inplace=True)

us_pop['Split'] = us_pop.Date.str.split(' ')
us_pop['Month'] = us_pop.Split.str.get(0)
us_pop['Year'] = us_pop.Split.str.get(1).astype(int) #Need to change type of year column to int to merge
us_pop.drop(columns=['Split', 'Date'], inplace=True)
us_pop.columns = ['US_Pop', 'Month', 'Year']

print(us_pop.head())
print(us_pop.describe())
print(us_pop.dtypes)
```

```python
dairy_df = pd.merge(left=production, right=us_pop, how='left', left_on=['Year', 'Month'], right_on=['Year', 'Month'])
#print(dairy_df.head())
dairy_df.info(null_counts=True)
```

```python
# Milk price units $/cwt
filepath = 'USPriceHistory/Milk-Time Series-Table 1.csv'
header = ['Date', 'Milk_Price']
milk_price = pd.read_csv(filepath, header=2, usecols=[0,1])
milk_price.columns = header
milk_price['Split'] = milk_price.Date.str.split(' ')
milk_price['Month'] = milk_price.Split.str.get(0)
milk_price['Year'] = milk_price.Split.str.get(1).astype(int) #Need to change type of year column to int to merge
milk_price.drop(columns=['Split', 'Date'], inplace=True)


print(milk_price.head())
print(milk_price.dtypes)
```

```python
# Corn Price units $/bu
filepath = 'USPriceHistory/US-Corn-Time Series-Table 1.csv'
header = ['Date', 'Corn_Price']
corn_price = pd.read_csv(filepath, header=2, usecols=[0,1])
corn_price.columns = header
corn_price['Split'] = corn_price.Date.str.split(' ')
corn_price['Month'] = corn_price.Split.str.get(0)
corn_price['Year'] = corn_price.Split.str.get(1).astype(int) #Need to change type of year column to int to merge
corn_price.drop(columns=['Split', 'Date'], inplace=True)

print(corn_price.head())
print(corn_price.dtypes)
```

```python
# Calve price in $/cwt
filepath = 'USPriceHistory/US-Calves-Time Series-Table 1.csv'
header = ['Date', 'Calves_price']
calves = pd.read_csv(filepath, header=2, usecols=[0,1])
calves.columns = header
calves['Split'] = calves.Date.str.split(' ')
calves['Month'] = calves.Split.str.get(0)
calves['Year'] = calves.Split.str.get(1).astype(int) #Need to change type of year column to int to merge
calves.drop(columns=['Split', 'Date'] , inplace=True)


print(calves.head())
print(calves.dtypes)
```

```python
# Cattle Price in $/cwt
filepath = 'USPriceHistory/US-Cattle-Time Series-Table 1.csv'
header = ['Date', 'Cattle_Price']
cattle = pd.read_csv(filepath, header=2, usecols=[0,1])
cattle.columns = header
cattle['Split'] = cattle.Date.str.split(' ')
cattle['Month'] = cattle.Split.str.get(0)
cattle['Year'] = cattle.Split.str.get(1).astype(int) #Need to change type of year column to int to merge
cattle.drop(columns=['Split', 'Date'], inplace=True)


print(cattle.head())
print(cattle.tail())
print(cattle.dtypes)
```

```python
# Soy Price in $/bu
filepath = 'USPriceHistory/US-Soybeans-Time Series-Table 1.csv'
header = ['Date', 'Soy_Price']
soy = pd.read_csv(filepath, header=2, usecols=[0,1])
soy.columns = header
soy['Split'] = soy.Date.str.split(' ')
soy['Month'] = soy.Split.str.get(0)
soy['Year'] = soy.Split.str.get(1).astype(int) #Need to change type of year column to int to merge
soy.drop(columns=['Split', 'Date'], inplace=True)


print(soy.head())
print(soy.dtypes)
```

```python
# Corn price in $/bu
filepath = 'USPriceHistory/US-Corn-Time Series-Table 1.csv'
header = ['Date', 'Corn_Price']
corn = pd.read_csv(filepath, header=2, usecols=[0,1])
corn.columns = header
corn['Split'] = corn.Date.str.split(' ')
corn['Month'] = corn.Split.str.get(0)
corn['Year'] = corn.Split.str.get(1).astype(int) #Need to change type of year column to int to merge
corn.drop(columns=['Split', 'Date'], inplace=True)


print(corn.head())
print(corn.dtypes)
```

```python
dairy_df = pd.merge(left=dairy_df, right=milk_price, how='left', left_on=['Year', 'Month'], right_on=['Year', 'Month'])
dairy_df = pd.merge(left=dairy_df, right=calves, how='left', left_on=['Year', 'Month'], right_on=['Year', 'Month'])
dairy_df = pd.merge(left=dairy_df, right=cattle, how='left', left_on=['Year', 'Month'], right_on=['Year', 'Month'])
dairy_df = pd.merge(left=dairy_df, right=soy, how='left', left_on=['Year', 'Month'], right_on=['Year', 'Month'])
dairy_df = pd.merge(left=dairy_df, right=corn, how='left', left_on=['Year', 'Month'], right_on=['Year', 'Month'])
print(dairy_df.head())
print(dairy_df.info(null_counts=True))
print(dairy_df.head())
```

```python
dairy_23 = dairy_df[dairy_df['Num_States'] == 23]
print(dairy_23.head())
print(dairy_23.info(null_counts=True))
```

```python
# Milk production vs. population growth
dairy_pop = dairy_df[dairy_df['Year'] >= 1990].dropna()

print(dairy_pop.head())
plt.plot(dairy_pop['US_Pop'], dairy_pop['Est_USA_Prod'])
plt.xlabel('US Population in Millions')
plt.ylabel('Est USA Production, in million lbs')
plt.show()
```

```python
# Plot milk per person
fig, ax = plt.subplots()
#https://matplotlib.org/gallery/api/date.html
years = mdates.YearLocator(5)   # every year
yearsFmt = mdates.DateFormatter('%Y')

dairy_pop['Milk_Per_Person'] = dairy_pop['Est_USA_Prod'] / dairy_pop['US_Pop']
dairy_pop['Date'] = dairy_pop['Year'].map(str) + ' ' + dairy_pop['Month']
print(dairy_pop.head())
print(dairy_pop.info())
dairy_pop['Date'] = pd.to_datetime(dairy_pop['Date'], format='%Y %B')

ax.plot(dairy_pop['Date'], dairy_pop['Milk_Per_Person']/8.6) #Convert milk qty to gallons

datemin = np.datetime64('1990-04')
datemax = np.datetime64('2018-09')
ax.set_xlim(datemin, datemax)

# format the ticks
ax.xaxis.set_major_locator(years)
ax.xaxis.set_major_formatter(yearsFmt)

plt.xlabel('Years')
plt.ylabel('Monthly Milk per person in gallons')


plt.show()
```
